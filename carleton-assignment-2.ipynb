```python
import pandas as pd
from datetime import datetime 
#####04.19.21 - IN ~9:50 am
##i'm going to be merging the csv version of the unconventional drilling permits sheet with the filtered, saved csv version of
###the zillow data, pa-2010s-only
### the two file names i'm joining are: zillow-PA-2010s-zip-data.csv & PA-Unconventional-Drilling-Permits_Issued_Detail (2).csv
```

```python
drillingpermits = pd.read_csv("PA-Unconventional-Drilling-Permits_Issued_Detail (2).csv")
##i'm importing the drilling permits dataset now|
## i'm retroactively assigning it to the name drillingpermits to make coding a little simpler
drillingpermits.head(10)
##i'm checking the head of the dataset to make sure all looks correct, and it does 
##i'm also making a mental note of the difference between the "ZIP-FULL" and the "Zip Code" columns in the below
##I used excel to remove the last 5 digits of "ZIP-FULL" into a new column with JUST the 5-digit zip code
##i kept both in the interest of keeping good track of my work, but I created the pared down Zip Code column because 
##that is what I will be joining the Zillow data on - the Zillow dataset also has a Zip Code column that's only 5-digits long 
```

```python
zillowdata = pd.read_csv("zillow-PA-2010s-zip-data.csv")
##and now i'm importing the filtered down zillow property value data 
###I imported the dataset with the pd.read_csv and now i'm going back and assigning it the 
###title 'zillowdata' 
zillowdata.head(6)
zillowdata.tail(6)
##i'm using the head and tail functions to get a sense of my dataset and make sure I've imported correctly
##it all looks correct, but there are null values at the bottom of the dataset. i double checked this against my CSV 
##and those are there correctly. If i had to guess, it's most likely because a zip code didn't have enough residences in it at the time for Zillow 
## to collect data on it. 
##also of note: the Zip Code column, spelled correctly w/ proper Caps, imported successfully which will make for an easy merge.
```

```python
##now i'm going to do an inner merge of the two datasets. I choose inner merge over other methods -- pd concat, outer merge, etc --
##because it is the method that will merge on one shared variable without creating any additional rows 
##this method will simply add the zip code data to the right of each drilling value, so I can see each drilling permit's location plus it's
##change in property value over time from 2010 to 2021
##i'm retroactively naming the merged file 'drillingzillow'
drillingzillow = drillingpermits.merge(zillowdata, how = "inner", on = "Zip Code")
```

```python
##i see from the above preview that my merge was successful! it has printed out a sheet with the correct number of rows
##with property value tacked on correctly to the right of all drill permit data
##i'm happy with this so i'm going to export the dataset now, just so I have it saved and don't lose my progress 
filename = "merged_drillingpermits-zillowvalues.csv"
drillingzillow.to_csv(filename, encoding= 'utf-8', index=False)
## I will return later to pandas to do some basic descriptive stats and get to know my data better on this platform
##anything I can't do here i will do in excel 
##just checked my assignment-2 folder on my desktop and a csv called 'merged_drillingpermits-zillowvalues' has arrived, hooray!
##04.19.21 - OUT 10:10 AM
```

```python
##now the next thing that i need to do is NORMALIZE the data and make a normalized-ten-year-change column to account for inflation
##so i need to make a 2010 normalized column, then resubtract. it's good to have a straight-up difference, but not accurate if not normalized. 
##i use the CPI calculator to see the difference between $100 in 2010 versus 2021 - https://www.bls.gov/data/inflation_calculator.htm 
##this tells me that $100 in Jan, 2010 would be $121.38 - so I know i need to multiply the 2010 column by 1.2138
##in order to see what property values in 2010 would be in 2021 dollars in order to compare the two


##here's me normalizing the year 2010: 
drillingzillow['jan-2010-normalized'] = drillingzillow['1/31/10']*1.2138
```

```python
drillingzillow.sample(10)
##this appears to have worked! now my next step is to do a normalized change over time column 
##by subtracting 2/28/21 minus jan-2010-normalized
```

```python
drillingzillow['ten-year-change-normalized'] = drillingzillow['2/28/21'] - drillingzillow['jan-2010-normalized']
```
